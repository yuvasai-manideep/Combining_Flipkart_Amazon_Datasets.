```python
# Loding Flipkart dataset.

flp_df=pd.read_csv("D:\downlods/flipkart_com-ecommerce_sample.csv")
```

```python
# Loding Amazon dataset.

amz_df=pd.read_csv("D:\downlods/amz_com-ecommerce_sample.csv",encoding = 'unicode_escape')
```

```python
#definig a function 

def find(a):
    #finding all pid's related to the product name.
    req=flp_df[flp_df["product_name"] == a][["pid"]]
    
    #creating a list with pid's
    list=[]
    for i in req["pid"]:
        list.append(i)
        
    #creating a DataFrame with the pid's in List.
    flp = flp_df[flp_df["pid"].isin(list)]
    amz = amz_df[amz_df["pid"].isin(list)]
    
    #filtering the dataframe with required columns.
    flp = flp[["product_name","retail_price","discounted_price","pid"]]
    amz = amz[["product_name","retail_price","discounted_price","pid"]]
    
    #merging dataframes.
    df = flp.merge(amz,left_on = "pid",right_on = "pid",how = "left").drop("pid",axis = 1)
    
    #Renaming the column_headers
    df = df.rename(columns = {"product_name_x":"Product name in Flipkart","product_name_y":"Product name in Amazon",
                          "retail_price_x":"Retail Price in Flipkart","retail_price_y":"Retail Price in Amazon",
                                "discounted_price_x":"Discounted Price in Flipkart",
                          "discounted_price_y":"Discounted Price in Amazon"})  
    return df
```

```python
def find_on(a,onn):
    
    flp = flp_df[flp_df[onn] == a]
    amz = amz_df[amz_df[onn] == a]
    
    #Slicing the dataframe
    flp = flp[["product_name","retail_price","discounted_price",onn]]
    amz = amz[["product_name","retail_price","discounted_price",onn]]
    
    #Merging the 2 DataFrames
    df = flp.merge(amz,left_on = onn ,right_on = onn ,how = "left").drop(onn,axis = 1)
    
    #renaming the column_headers
    df = df.rename(columns = {"product_name_x":"Product name in Flipkart","product_name_y":"Product name in Amazon",
                          "retail_price_x":"Retail Price in Flipkart","retail_price_y":"Retail Price in Amazon",
                                "discounted_price_x":"Discounted Price in Flipkart",
                          "discounted_price_y":"Discounted Price in Amazon"})  
    return df
```
